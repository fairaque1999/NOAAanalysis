```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
from sklearn.cluster import KMeans
```

```python
data = pd.read_csv('/Users/ishraque/Documents/Mars_Glacier_Boulders/site_a.csv', sep=',')
```

```python
plt.scatter(data['x'], data['y2'])
```

```python
new_data = data[['x','y2']]
Kmean1 = KMeans(n_clusters=5)
Kmean1.fit(new_data)
y_kmeans = Kmean1.predict(new_data)
```

```python
x = Kmean1.cluster_centers_
```

```python
plt.scatter(new_data['x'],new_data['y2'], c=y_kmeans, s=50, cmap='viridis')
for xc in x[:,0]:
    plt.axvline(x=xc, color='black', linestyle='--')
```

```python
boulderdata = pd.read_csv('/Users/ishraque/Documents/Mars_Glacier_Boulders/Glacier Boulder Data/A_boulders_1.txt', sep=',')
boulderdata.head()
```

```python
newboulder = boulderdata[['Northing', 'Easting']]
```

```python
Kmean2 = KMeans(n_clusters=50)
Kmean2.fit(newboulder)
y_kmeans = Kmean2.predict(newboulder)
x2 = Kmean2.cluster_centers_
```

```python
plt.scatter(newboulder['Northing'],newboulder['Easting'], c=y_kmeans, s=50, cmap='viridis')
for xn in x2[:,0]:
    plt.axvline(x=xn, color='black', linestyle='--')
```
